```python
#2 & 3. Create and label 4 separate charts with the following equations (choose a range for x that makes sense):


ax fig, (ax0, ax1, ax2, ax3) = plt.subplots(nrows=4,
                                 ncols=1,
                                 sharex=False, 
                               figsize = (10,8))
x_axis=[]
y_axis=[]
for x in range(0, 50):
    y = math.sqrt(x)
    x_axis.append(x)
    y_axis.append(y)
    
ax0.plot(x_axis, y_axis, c='purple', label = 'sqrt y')
ax0.set_title('sqrt y')

x_axis=[]
y_axis=[]
for x in range(0, 25):
    y = x**3
    x_axis.append(x)
    y_axis.append(y)
    
ax1.plot(x_axis, y_axis, c='red', label = 'y = x^3')
ax1.set_title('y = x^3')

x_axis=[]
y_axis=[]
for x in range(0, 50):
    y = 2**x
    x_axis.append(x)
    y_axis.append(y)
    
ax2.plot(x_axis, y_axis, c='turquoise', label = 'y = 2 ** 2')
ax2.set_title('y = 2 ** 2')

x_axis=[]
y_axis=[]
for x in range(0, 25):
    y = 1/(x+1)
    x_axis.append(x)
    y_axis.append(y)
    
ax3.plot(x_axis, y_axis, c='limegreen', label = ' y = 1\(x + 1')
ax3.set_title('y = 1\(x + 1)')
```
