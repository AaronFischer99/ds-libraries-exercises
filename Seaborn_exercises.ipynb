```python
#orders.groupby(['item_name'])['price'].sum()
df5 = orders.groupby(['item_name', 'quantity']).agg([sum]).drop(['item_price', 'choice_description', 'order_id', 'id'], axis=1)
#pop_items = orders.groupby('item_name')['quantity'].sum()
#.nlargest()
review = orders.groupby('item_name')['price'].sum().nlargest(4)
```

```python
df5.groupby('item_name')['price'].sum().nlargest(4)
```

```python
orders = orders.assign(price = orders['item_price'].str.replace('$','').astype(float))
```

```python
#review = orders.groupby('item_name')['price'].sum().nlargest(4)
revenue = orders.groupby('item_name')['price'].sum().nlargest(10)

revenue
```

```python
pop_items = orders.groupby('item_name')['quantity'].sum().nlargest(10)
pop_items
```

```python
top_chipotle_items = pd.merge(pop_items, revenue,on = 'item_name', how = 'inner')
top_chipotle_items.head(4)
```
